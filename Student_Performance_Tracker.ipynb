```python
class Student:
  def __init__(self):
    self.name = self.only_str("Enter Student's Name  : ")
    self.age = self.only_int("Enter Student's Age  : ")
    self.gender = self.only_str("Enter Student's Gender  : ")
    self.roll_no = self.only_int("Enter Student's Roll No  : ")
    self.score_in_maths = self.only_score(f"Enter {self.name}'s Score in Maths Out of 100 ")
    self.score_in_science = self.only_score(f"Enter {self.name}'s Score in Science Out of 100 ")
    self.score_in_english = self.only_score(f"Enter {self.name}'s Score in English Out of 100 ")
    self.score_in_chemistry = self.only_score(f"Enter {self.name}'s Score in Chemistry Out of 100 ")
    self.score_in_computer = self.only_score(f"Enter {self.name}'s Score in Computer Out of 100 ")
    self.score_in_physics = self.only_score(f"Enter {self.name}'s Score in Physics Out of 100 ")

  def __str__(self):
    return ( f"The name of the student is {self.name} , Age Is {self.age} , Gender Is {self.gender} and Roll number Is {self.roll_no}\n"
          f"Score of {self.name} is as follows\n"
          f"Math = {self.score_in_maths}, Science = {self.score_in_science}, English = {self.score_in_english} ,Chemistry = {self.score_in_chemistry} ,Computer = {self.score_in_computer} ,Physics = {self.score_in_physics}\n"
          f"The Average Score Of {self.name} Is {self.calculate_average()}\n"
          f"The Detailed Academic Performance of {self.name} Is as follows\n{self.is_passing()}\n"
           )

  def only_str(self,input_str):
    while True:
      valid_input = True
      try:
        input_from_user = input(input_str).strip()
        if not input_from_user:
          valid_input = False
          raise ValueError
        for char in input_from_user:
          if not(char.isalpha() or char.isspace()):
            valid_input = False
            raise ValueError
        if valid_input:
            return input_from_user
      except ValueError:
        print("Invalid Input, Please Enter In Alphabatical Letters Only And No Empty Input ")

  def only_int(self,input_int):
      while True:
        try:
          input_from_user = int(input(input_int))
          return input_from_user
        except ValueError:
          print("Invalid Input, Please Enter in Integers Only And No Empty Input ")

  def only_score(self,input_score):
    while True:
      try :
        input_from_user = int(input(input_score))
        if 0 <= input_from_user <= 100:
          return input_from_user
        else:
          print("Invalid Input, Please Enter the Valid Score Between 0 And 100 ")
      except ValueError:
        print("Invalid Input, Please Enter in Integers Only And No Empty Input ")

  def calculate_average(self):
    self.average = (self.score_in_maths + self.score_in_science + self.score_in_english + self.score_in_chemistry + self.score_in_computer + self.score_in_physics)/6
    return self.average

  def is_passing(self):
    result = []
    if self.score_in_maths >= 40 :
      result.append(f"{self.name} Has Passed In Maths")
    else :
      result.append(f"{self.name} Has Failed In Maths")
    if self.score_in_science  >= 40 :
       result.append(f"{self.name} Has Passed In Science")
    else :
      result.append(f"{self.name} Has Failed In Science")
    if self.score_in_english >= 40 :
       result.append(f"{self.name} Has Passed In English")
    else :
      result.append(f"{self.name} Has Failed In English")
    if self.score_in_chemistry >= 40 :
       result.append(f"{self.name} Has Passed In Chemistry")
    else :
      result.append(f"{self.name} Has Failed In Chemistry")
    if self.score_in_computer >= 40 :
       result.append(f"{self.name} Has Passed In Computer")
    else :
      result.append(f"{self.name} Has Failed In Computer")
    if self.score_in_physics >= 40 :
       result.append(f"{self.name} Has Passed In Physics")
    else :
      result.append(f"{self.name} Has Failed In Physics")
    results = "\n".join(result)
    return(results)
```

```python
class Performance_tracker:
  def __init__(self,student_dict):
    self.student_dict = student_dict

  def add_student(self):
    while True:
      student = Student()
      self.student_dict[student.name] = student
      print(f"Added {student.name} to the Performance Tracker ")
      while True:
        add_more = input("Do you want to Add More Students to the Performance Tracker  Yes/No ").lower()
        if add_more == "no":
          return
        if add_more == "yes":
          break
        else:
          print("Invalid Input , Please Give Your Answer in Yes Or No ")
  def display_students_info(self):
    if not self.student_dict:
            print("No students in the Performance Tracker.")
    else:
        print("\nCurrent Students in the Performance Tracker:")
        for name, student in self.student_dict.items():
           print(student)

  def calculate_class_average(self):
    if not self.student_dict:
      print("No Students In Performance Tracker")
      return
    else:
      total_average = 0
      for student in self.student_dict.values():
        total_average += student.calculate_average()
        class_average = total_average / len(self.student_dict)
      return f"The Average Of The Whole Class Is {class_average:.2f}"




```
